```python
def add(a,b,c):
    return a+b+c
def sub(a,b,c):
    return a-b-c
def mul(a,b,c):
    return a*b*c
def dvd(a,b,c):
    return a/b/c

def cal():
    a = int(input("Enter the value of a = "))
    b = int(input("Enter the value of b = "))
    c = int(input("Enter the value of c = "))
    o = input("Enter the operators = ")
    if o == '+':
        print(add(a,b,c))
    elif o == '-':
        print(sub(a,b,c))
    elif o == '*':
        print(mul(a,b,c))
    elif o == '/':
        print(dvd(a,b,c))
    else:
        print("Sorry no Calculation could be performed")
    
    x = input("Enter Y for more Calculations = ")
    if x == 'y':
        cal()
cal()
```

```python
bill = ""
n = int(input("Enter the n = "))
def billing():
    product = input("Enter the name of product = ")
    price = int(input("Enter the price = "))
    quantity = int(input("Enter the quantity = "))
    total = price*quantity
    info = f"{product} {price} {quantity} {total}\n"
    bill = bill + info
    for i in range(n):
        billing()
print(bill)
```
